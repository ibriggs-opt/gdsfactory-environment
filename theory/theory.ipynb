```python
import numpy as np
import matplotlib.pyplot as plt

ng = 4.2 #group index of silicon waveguide
dL = np.linspace(500,1.1e3,10000) #um, length difference between upper and lower arms
wavelength = 1.55 #um

fsr = (wavelength**2)/(ng*dL)

plt.plot(dL,fsr*(1e3))
plt.xlabel('dL (um)')
plt.ylabel('FSR (nm)')
plt.show()
```
